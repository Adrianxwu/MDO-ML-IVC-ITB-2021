- Import raw data from the source.

Usually, the data preprocessing steps also includes the data cleansing, manipulation, and transformation. However, because the given data is already 'clean' such steps is not needed.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def data_preprocessing(filepath, header, target_label, test_size = 0.2, random_state = 3):
    """
    A simple data preprocessing function.

    Args:
        filepath (str): file path string.
        header (list): list of header name
        target_label (str): target label header
        test_size (float): test size fraction, default to be 0.2
        random_state (int): random seed, default to be 3
    
    Returns:
        preprocessed_data (list): A list that contains Xtrain, Xtest, ytrain, ytest 
    """
    # Import data and give header
    df = pd.read_csv(filepath, sep='\t', names=header)

    # Separate target label
    X = df.drop(columns=[target_label])
    y_ref = df[target_label].values.reshape(X.shape[0], 1)

    # Split data
    Xtrain, Xtest, ytrain, ytest = train_test_split(X, y_ref, test_size=test_size, 
                                                    random_state=random_state)
    
    # Standardize X
    sc = StandardScaler()
    sc.fit(Xtrain)
    Xtrain = sc.transform(Xtrain)
    Xtest = sc.transform(Xtest)

    # Standardize y
    # sc2 = StandardScaler()
    # sc2.fit(ytrain)
    # ytrain = sc2.transform(ytrain)
    # ytest = sc2.transform(ytest)

    return [Xtrain, Xtest, ytrain, ytest]
```

#### Weight and Biases in Neural Network

* In the previous section we've learned about the intuitive of weights and biases. 
* However, when we the neurons are stacked, the "meaning" of weights and biases will become less interpretable and less meaningful to us.
* Hence, we can treat those weights and biases as parameters that will help our model to fit the given data.

So, here comes another problem: "How do we compute the weight and biases in Python?"
* We could use matrices!
* Recall our neuron's weighted sum equation:

$$
y = w_1 * x_1 + w_2*x_2 + \ldots + w_n*x_n + b
$$

* This can be rewritten as:

$$
y = \begin{bmatrix}
x_1 & x_2 & \ldots & x_n
\end{bmatrix}
\cdot
\begin{bmatrix}
w_1 \\ w_2 \\ \ldots \\ w_n
\end{bmatrix}
+ b
$$

* That's for a single neuron. Now, let's pretend that we want to calculate the weights and bias for the hidden layer in this network.

<img src="https://i.stack.imgur.com/mjeXO.png" width= 400px;/>

[source](https://i.stack.imgur.com/mjeXO.png)

* In the hidden layer, we have 2 neurons with 3 inputs. Hence, it could be written as:

$$
y = \begin{bmatrix}
x_1 & x_2 & x_3
\end{bmatrix}
\cdot
\begin{bmatrix}
w_{11} & w_{12} \\ w_{21} & w_{22} \\ w_{31} & w_{32}
\end{bmatrix}
+
\begin{bmatrix}
b_{1} & b_{2}
\end{bmatrix}
$$

* Now, we noticed that the weight matirx has the shape of $3 \times 2$ or, in general $prev\_layer \times current\_layer$.
* Hence, if you have 2 layers architecture such as $[4,5,3,1]$, you'll have 3 weights matrices and 3 biases with the following shapes:
    * $W_1 = (4,5)$, $b_1 = (5,)$
    * $W_2 = (5,3)$, $b_2 = (3,)$
    * $W_3 = (3,1)$, $b_1 = (1,)$

#### Code

Now, let's create a simple Python Class for our Neural Network weights and biases. Later, this code will be updated along the tutorial

```python
import numpy as np

class NN():
    """
    A simple two layers NN
    """

    def __init__(self, layers=[5,8,1]):
        # variable defense, we don't want less or more layers
        assert(len(layers) == 3), "NN has to be 2 layers."
        
        self.params = {}
        self.layers = layers
        self.X = None
        self.y = None
    
    def init_weights(self, random_seed = 1):
        """
        Weights and biases initialization from random normal distribution
        """
        np.random.seed(random_seed) # Seed the random number generator
        self.params["W1"] = np.random.randn(self.layers[0], self.layers[1]) 
        self.params['b1'] = np.random.randn(self.layers[1],)
        self.params['W2'] = np.random.randn(self.layers[1],self.layers[2]) 
        self.params['b2'] = np.random.randn(self.layers[2],)
```

### Loss Function

Now, let's talk about Loss Function. What is a loss function?
* A function.
* Measure how good a model's prediction.

When we builds a ML/Statistical model, there are a lot of loss functions that can be used depends on the case and model. Maybe, for some of you that already quite familiar with OLS, you have heard about Mean Squared Error (MSE) or Mean Absolute Error (MAE).

Those two are some example of loss function.In this tutorial, as we have regression problem, we will use the Mean Squared Error (MSE).



#### Mean Squared Error Loss Function

The general equation for MSE loss :

$$
L_{MSE} = - \frac{1}{n} \sum_i^n (y_i -\hat{y_i})^2
$$


#### Code


```python
def mse (y, yhat):
    nsamp = len(y)
    loss = 1/nsamp * (np.sum((y-yhat)**2))
    return loss
```

### Forward Propagation

Up to this point, we have some basic building blocks for our Neural Network code. Now, we'll put everything together and performs a process called forward propagation.

Basically, in forward propagation we want to perform a series of computations in the neural network from the input layer to the output layer **without tuning the parameters**:
* Compute the weighted sum between input and the first layer.
* Pass the result through the 1st activation function.
* Computer the weighted sum between the previous layer and the next layer.
* Pass the result through the next activation function.
* Repeat until the output layer is reached.
* Finally, compute the loss between the output and the true labels.

Now, we'll put everything's together and create a new method in the `NN` class: 

```python
class NN():
    """
    A simple two layers NN
    """

    def __init__(self, layers=[5,8,1]):
        # variable defense, we don't want less or more layers
        assert(len(layers) == 3), "NN has to be 2 layers."
        
        self.params = {}
        self.layers = layers
        self.X = None
        self.y = None
        self.loss = []
    
    def init_weights(self, random_seed = 1):
        """
        Weights and biases initialization from random normal distribution
        """
        np.random.seed(random_seed) # Seed the random number generator
        self.params["W1"] = np.random.randn(self.layers[0], self.layers[1]) 
        self.params['b1'] = np.random.randn(self.layers[1],)
        self.params['W2'] = np.random.randn(self.layers[1],self.layers[2]) 
        self.params['b2'] = np.random.randn(self.layers[2],)
    
    def relu(self, z):
        """
        relu function will return z if z > 0. Otherwise, 0
        """
        return np.maximum(0,z)
    
    def mse (self, y, yhat):
        nsamp = len(y)
        loss = 1/nsamp * (np.sum((y-yhat)**2))
        return loss
    
    def forward_prop (self):
        Z1 = self.X.dot(self.params['W1']) + self.params['b1']
        A1 = self.relu(Z1)
        Z2 = A1.dot(self.params['W2']) + self.params['b2']
        yhat = Z2
        loss = self.mse(self.y,yhat)

        # save calculated parameters     
        self.params['Z1'] = Z1
        self.params['Z2'] = Z2
        self.params['A1'] = A1

        return yhat,loss
```

### Backpropagation

* Process in neural network to update its weights and biases.
* We want to update the weights and biases until it reached the minimum loss.
* Basically, this is an optimization problem.
* We want to adjust NN weights with respect to the loss effectively.
* We want to find the derivative of the loss function with respect to each variables!

Now, we've created a new problem. How do we calculate the derivatives?
- We can use the **chain rule**!

#### Chain Rule Refresher

Chain rule is a formula to compute the derivative of a composite function.

For example:
* $z$ is a function that depends on variable $y$.
* But $y$ itself is a variable that depends on $x$.
* Then $z$, via the intermediate variable $y$, depends on $x$ as well.
* The chain rule can be written as:

$$
\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx}
$$

#### NN Back Propagation

> **Disclaimer**: Since the time is limited, we won't derive the derivatives from scratch. If you are interested, this [post](https://medium.com/@pdquant/all-the-backpropagation-derivatives-d5275f727f60) is relatively easy to read.

Now, as the name implies, the backprop starts from the last layer:
* First, we want to calculate the derivative of the loss with respect to the output *y_hat*

$$
dl\_wrt\_yhat = 2(\hat{y} - y) 
$$

* Now, how did we get the $\hat{y}$. We calculated it through the linear function

$$
dlinear = 1
$$

* Now, we calculate the derivative of the loss wrt Z2:

$$
dl\_wrt\_Z2 = dl\_wrt\_yhat \times dlinear
$$

* Because the derivative of linear function is 1

$$
dl\_wrt\_Z2 = dl\_wrt\_yhat
$$

* Repeat all of the process with the same pattern:

$$
dl\_wrt\_A1 = W2 * dl\_wrt\_Z2
$$

$$
dl\_wrt\_W2 = A1 * dl\_wrt\_Z2
$$

$$
dl\_wrt\_b2 = A1 * dl\_wrt\_Z2
$$

* The derivative of ReLU is 1 if the input is greater than 1, and 0 otherwise. We'll create a new function for this.
* Repeat the same process:

$$
dl\_wrt\_Z1 = dl\_wrt\_A1 * dRelu
$$

$$
dl\_wrt\_W1 = X * dl\_wrt\_Z1
$$

$$
dl\_wrt\_b1 = dl\_wrt\_Z1
$$

### Training Process

* In order to get the optimum weights and biases, NN has to go through many forward and backward propagation cycle.
* This cycle is what we generally refer to the training process.

So, in general training process consist of:
* Forward prop
* Backprop
* Weights and biases updates
* Repeat

Now, the problem is. How do we update the Weights with known gradients from backprop?
* Introducing `learning_rate`.
* If we assume that the gradient is the slope of a mountain.
* Learning rate is a variable that decides "how big the step that we want to take".
* If the step is too small, then we would take a long time.
* If the step is too big, we would fall and die.

putting the backprop and training process all together:

```python
def dRelu(self, x):
        x[x<=0] = 0
        x[x>0] = 1
        return x

def back_propagation(self,yhat):

        dl_wrt_yhat = 2*(yhat - self.y)
        dl_wrt_z2 = dl_wrt_yhat

        dl_wrt_A1 = dl_wrt_z2.dot(self.params['W2'].T)
        dl_wrt_w2 = self.params['A1'].T.dot(dl_wrt_z2)
        dl_wrt_b2 = np.sum(dl_wrt_z2, axis=0, keepdims=True)

        dl_wrt_z1 = dl_wrt_A1 * self.dRelu(self.params['Z1'])
        dl_wrt_w1 = self.X.T.dot(dl_wrt_z1)
        dl_wrt_b1 = np.sum(dl_wrt_z1, axis=0, keepdims=True)

        # Updates weights
        self.params['W1'] = self.params['W1'] - self.learning_rate * dl_wrt_w1
        self.params['W2'] = self.params['W2'] - self.learning_rate * dl_wrt_w2
        self.params['b1'] = self.params['b1'] - self.learning_rate * dl_wrt_b1
        self.params['b2'] = self.params['b2'] - self.learning_rate * dl_wrt_b2
```

Also, we want to add a function to train the model:

```python
def fit(self, X, y):
        '''
        Trains the neural network using the specified data and labels
        '''
        self.X = X
        self.y = y
        self.init_weights() #initialize weights and bias


        for i in range(self.iterations):
            yhat, loss = self.forward_propagation()
            self.back_propagation(yhat)
            self.loss.append(loss)
```

### Making Prediction

Making a prediction is rather straightforward. We want to pass the data through the trained NN model. Then, because this is a classification task, we want to make that if the predicted probability is more than 0.5, then it should belong to class 1, otherwise 0. We can achieve this through `np.round` 

```python
def predict(self, X):
        '''
        Predicts on a test data
        '''
        Z1 = X.dot(self.params['W1']) + self.params['b1']
        A1 = self.relu(Z1)
        Z2 = A1.dot(self.params['W2']) + self.params['b2']
        pred = Z2
        return pred 
```

### Put Everything Together

```python
import matplotlib.pyplot as plt

class NN():
    """
    A simple two layers NN
    """

    def __init__(self, layers=[5,8,1], learning_rate=0.001, iterations=100):
        # variable defense, we don't want less or more layers
        assert(len(layers) == 3), "NN has to be 2 layers."
        
        self.params = {}
        self.layers = layers
        self.X = None
        self.y = None
        self.loss = []
        self.learning_rate = learning_rate
        self.iterations = iterations
    
    def init_weights(self, random_seed = 1):
        """
        Weights and biases initialization from random normal distribution
        """
        np.random.seed(random_seed) # Seed the random number generator
        self.params["W1"] = np.random.randn(self.layers[0], self.layers[1]) 
        self.params['b1'] = np.random.randn(self.layers[1],)
        self.params['W2'] = np.random.randn(self.layers[1],self.layers[2]) 
        self.params['b2'] = np.random.randn(self.layers[2],)
    
    def relu(self, z):
        """
        relu function will return z if z > 0. Otherwise, 0
        """
        return np.maximum(0,z)
    
    def dRelu(self, x):
        x[x<=0] = 0
        x[x>0] = 1
        return x
    
    def mse (self, y, yhat):
        nsamp = len(y)
        loss = 1/(nsamp) * (np.sum(np.square(y-yhat)))
        return loss

    def nmse (self, y, yhat):
        nsamp = len(y)
        loss = 1/(nsamp) * (np.sum(np.square(y-yhat))) * 1/(np.sum(np.square(y-np.mean(y))))
        return loss
    
    def forward_prop (self):
        Z1 = self.X.dot(self.params['W1']) + self.params['b1']
        A1 = self.relu(Z1)
        Z2 = A1.dot(self.params['W2']) + self.params['b2']
        yhat = Z2
        loss = self.mse(self.y,yhat)

        # save calculated parameters     
        self.params['Z1'] = Z1
        self.params['Z2'] = Z2
        self.params['A1'] = A1

        return yhat,loss
    
    def back_propagation(self,yhat):
        
        dl_wrt_yhat = 2/len(yhat) * (yhat - self.y)
        dl_wrt_z2 = dl_wrt_yhat

        dl_wrt_A1 = 2/len(yhat) * dl_wrt_z2.dot(self.params['W2'].T)
        dl_wrt_w2 = self.params['A1'].T.dot(dl_wrt_z2)
        dl_wrt_b2 = 2/len(yhat) * np.sum(dl_wrt_z2, axis=0, keepdims=True)

        dl_wrt_z1 = 2/len(yhat) * dl_wrt_A1 * self.dRelu(self.params['Z1'])
        dl_wrt_w1 = self.X.T.dot(dl_wrt_z1)
        dl_wrt_b1 = 2/len(yhat) * np.sum(dl_wrt_z1, axis=0, keepdims=True)

        # Updates weights
        self.params['W1'] = self.params['W1'] - self.learning_rate * dl_wrt_w1
        self.params['W2'] = self.params['W2'] - self.learning_rate * dl_wrt_w2
        self.params['b1'] = self.params['b1'] - self.learning_rate * dl_wrt_b1
        self.params['b2'] = self.params['b2'] - self.learning_rate * dl_wrt_b2
    
    def fit(self, X, y):
        '''
        Trains the neural network using the specified data and labels
        '''
        self.X = X
        self.y = y
        self.init_weights() #initialize weights and bias


        for i in range(self.iterations):
            yhat, loss = self.forward_prop()
            self.back_propagation(yhat)
            self.loss.append(loss)
    
    def predict(self, X):
        '''
        Predicts on a test data
        '''
        Z1 = X.dot(self.params['W1']) + self.params['b1']
        A1 = self.relu(Z1)
        Z2 = A1.dot(self.params['W2']) + self.params['b2']
        pred = Z2
        return pred
    
    def plot_loss(self):
        '''
        Plots the loss curve
        '''
        plt.plot(self.loss)
        plt.xlabel("Iteration")
        plt.ylabel("MSE")
        plt.title("Loss curve for training")
        plt.show()
```

#### 2 Variable Function

Given a function ([Booth Function](https://www.sfu.ca/~ssurjano/booth.html)):
$$
f(x,y) = (x + 2y -7)^2 + (2x+y-5)^2
$$

Where $-10\leq x,y \leq 10 $

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

booth = lambda x: (x[:,0] + 2*x[:,1] - 7)**2 + (2*x[:,0] + x[:,1] - 5)**2

def generate_sample(func,bound=[-10,10], n=100):
    """
    Function to generate samples 
    Args:
        - func(callable): Evaluation function
        - bound(list): minimum and maximum bound.
        - n(int): number of grid in one axis
    
    Return:
        - Xeval(np array)
        - yeval
    """

    xx = np.linspace(bound[0], bound[1], n)
    yy = np.linspace(bound[0], bound[1], n)
    Xevalx, Xevaly = np.meshgrid(xx, yy)
    Xeval = np.zeros(shape=[n**2, 2])
    Xeval[:, 0] = np.reshape(Xevalx, (n**2))
    Xeval[:, 1] = np.reshape(Xevaly, (n**2))

    yeval = func(Xeval)

    return Xeval, yeval

def surf_plot(X, y):
    n = int(np.sqrt(len(y)))
    y = y.reshape(n,n)
    x1 = X[:,0].reshape(n,n)
    x2 = X[:,1].reshape(n,n)

    fig = plt.figure()
    ax = fig.gca(projection="3d")
    surf = ax.plot_surface(
        x1, x2, y, cmap=cm.coolwarm, linewidth=0, antialiased=False
    )
```

```python
X,y = generate_sample(booth)
X_samp, y_samp = generate_sample(booth, n=10)
surf_plot(X,y)

fig = plt.figure()
ax = fig.gca(projection="3d")
surf = ax.scatter(
    X_samp[:,0], X_samp[:,1], y_samp
)

plt.show()
```

```python
x_samp_norm = X_samp / 10

nn = NN([2,16,1], learning_rate=0.01, iterations=100000)
nn.fit(x_samp_norm, y_samp.reshape(-1,1))
```

```python
nn.plot_loss()
```

```python
pred = nn.predict(X/10)
```

```python
surf_plot(X, pred)
```

### Try using PyTorch

```python
import torch
from torch.autograd import Variable
import torch.nn.functional as F
```

```python
x, y = torch.from_numpy(Xtrain).float(), torch.from_numpy(ytrain).float() 
x, y = Variable(x), Variable(y)
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)   # hidden layer
        self.predict = torch.nn.Linear(n_hidden, n_output)   # output layer

    def forward(self, x):
        x = F.relu(self.hidden(x))      # activation function for hidden layer
        x = self.predict(x)             # linear output
        return x

net = Net(n_feature=5, n_hidden=8, n_output=1)     # define the network
# print(net)  # net architecture
optimizer = torch.optim.SGD(net.parameters(), lr=0.001)
loss_func = torch.nn.MSELoss()  # this is for regression mean squared loss

# train the network
train_losses = []
for t in range(100):
  
    prediction = net(x)     # input x and predict based on x

    loss = loss_func(prediction, y)     # must be (1. nn output, 2. target)
    train_losses.append(loss.item())

    optimizer.zero_grad()   # clear gradients for next train
    loss.backward()         # backpropagation, compute gradients
    optimizer.step()        # apply gradients

plt.plot(train_losses)
```
