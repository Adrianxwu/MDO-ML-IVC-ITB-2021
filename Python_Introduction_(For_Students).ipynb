## Dictionary
Dictionary is a collection of data values and its keypair.

````python
Dictionary_data = {"James": 35, "Thomas": 40, "Jessica": 32, "Nicholas": 27}

print(Dictionary_data["Jessica"])
````

````python
Dictionary_data = {"James": {"Age": 35, "Country": "USA"}, "Thomas": {"Age": 40, "Country": "HK"}, "Jessica": {"Age": 32, "Country": "UK"}}

print(Dictionary_data["Jessica"])
print(Dictionary_data["Jessica"]["Country"])
````

To add an entry to an existing dictionary, we can do the following:

````python
Dictionary_data = {"Thomas": {"Age": 40, "Country": "HK"}}

Dictionary_data["James"] = {"Age": 35, "Country": "USA"}

print(Dictionary_data)
````

To delete an element from a dictionary, we can use the "del" function

````python
Dictionary_data = {"James": {"Age": 35, "Country": "USA"}, "Thomas": {"Age": 40, "Country": "HK"}, "Jessica": {"Age": 32, "Country": "UK"}}

del Dictionary_data["Thomas"]

print(Dictionary_data)
````

In previous version of Pythons, dictionary is unordered and the data insertion order is not perserved. However, in Python version 3.6 and above, dictionary is ordered and the data insertion order is perserved.

````python
Dictionary_data = {"a":0, "b":1, "c":2, "d":3, "e":4}
Dictionary_data["f"] = 5
Dictionary_data["g"] = 6
Dictionary_data["h"] = 7
Dictionary_data["i"] = 8

print(Dictionary_data)

# Because dictionary is ordered now, we can safely access the data using the insertion order
print(list(Dictionary_data.items())[4]) # "list(Dictionary_data.items())" creates a list version of the dictionary
````

## Conditions
Python uses boolean variable consisting of 2 possible values(true and false) to evaluate conditions.


````python
a = 10
print(a==2) # "==" operator is the "is equal?" boolean operator
print(a==10)
print(a!=10) # "!=" operator is the "not equal?" boolean operator
````

The "in" operator is a boolean operator that can be used to check whether a specified object exists in a list.

````python
a = 20
b = 10
list_data = [1,5,6,7,20,25]

print(a in list_data)
print(b in list_data)
````

Boolean variables can be used in if conditional expression. In the following code snippets, the code first evaluates the first "if" conditional and if it returns true, the indented block of code bellow is executed. Otherwise, it proceeds to the next "elif" (else if) conditional and so on.

````python
a = 12
if (a == 5):
    print("a is equal to 5") # This indented part of the code is executed if (a==5) returns true
elif (a == 10): # "elif" is else if in Python
    print("a is equal to 10")
else:
    print("a is not equal to 5 or 10")
````

## Loops
In Python, there are 2 types of loops, "for" and "while". "For" loops iterate over a given sequence and "while" loops repeats until a certain boolean expression returns false. Example of "for" loops:

````python
list_data = []

for i in range(10):
    list_data.append(i)

print(list_data)
````

````python
list_data = []

for i in range(5,10):
    list_data.append(i)

print(list_data)
````

````python
list_data = []

for i in range(10,5,-1):
    list_data.append(i)

print(list_data)
````

````python
list_data = ["hello",1,2,3,"world"]

for i in list_data:
    print(i)
````

Example of "while" loops:

````python
a = 5
while a < 10:
    print(a)
    a = a + 1
````

"Break" statement can be used to break a loop process and continue the code outside the loop.

````python
for i in range(5):
    print(i)
    if (i == 2):
        break
````

"Continue" statement can be used to skip certain parts of a code inside a loop

````python
for i in range(5):
    if (i == 2):
        continue
    print(i)
````

````python
def addition_function(a,b):
    result = a + b
    return result
````

##Initializing Numpy Array

## Accessing and filling existing Numpy array

````python
# Initializing zeros array
data = np.zeros((5,3))
print("initalized array:")
print(data)

# Filling the zeros array
data[1,0] = 1
data[0,1] = 2
data[4,1] = 3
data[3,2] = 4
print("filled array:")
print(data)

# Accessing individual element of the filled array
data_4_1 = data[4,1]
print("Element (4,1):")
print(data_4_1)
````

##Matrix Operations in Numpy

````python
# Creating Matrix
A = np.array([[1,2,3],[3,2,5],[2,1,-4]])
B = np.array([[1,1,2],[1,5,6],[0,5,6]])
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)

# Matrix Multiplication
C = np.matmul(A,B)
print("A x B:")
print(C)

# Matrix Determinant
detA = np.linalg.det(A)
print("Determinant of A:")
print(detA)

# Matrix Inverse
invA = np.linalg.inv(A)
print("Inverse of A:")
print(invA)
````

We can also visualize 2D function's contour using this scatter plot

````python
import numpy as np

# Distributing grid and calculating f = x^2 + y^2 at each grid point
x = []
y = []
f = []

for i in range(100):
    for j in range(100):
        x.append(0.1*i)
        y.append(0.1*j)
        f.append((0.1*i)**2 + (0.1*j)**2)

# Scatter Plot
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis("equal")
plt.scatter(x,y,c=f,s=5)
plt.colorbar()
````

## Contour Plot

To use the built in contour plotting function in Matplotlib, we need to learn about "meshgrid" first. For 2-D, Meshgrid creates a rectangular grid from 2 1-D arrays (x and y).

````python
x = np.linspace(1,4,4)
y = np.linspace(1,5,5)

X,Y = np.meshgrid(x,y)
print("x:")
print(x)
print("X:")
print(X)
print("y:")
print(y)
print("Y:")
print(Y)
````

````python
import numpy as np

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
X,Y = np.meshgrid(x,y)
F = X**2 + Y**2

# Contour Plot
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis("equal")
plt.contourf(X,Y,F,10) # We can increase the value-steps to smoothen the contour
plt.colorbar()
````
